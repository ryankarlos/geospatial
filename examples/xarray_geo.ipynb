```python
%matplotlib inline
import numpy as np
import pandas as pd
import xarray as xr
from matplotlib import pyplot as plt
```

```python

ds = xr.tutorial.open_dataset('rasm').load()
ds.Tair.isel(time=5).sel(x=slice(0,None), y =slice(0, None)).plot()
```

```python
monthly_mean = ds.Tair.resample(time = 'M').mean()
monthly_mean.sel(x=slice(100, None), y =slice(200, None)).plot()
```

```python
ds.Tair.groupby('time').mean(dim='x').plot()
```
